```python
import pandas as pd

X_train = pd.read_csv('../data/X_train.csv', parse_dates=['reportts'])
y_train = pd.read_csv('../data/y_train.csv', parse_dates=['reportts'])

X_test = pd.read_csv('../data/X_test.csv', parse_dates=['reportts']).fillna(0)

X_train.rename(columns={"reportts": "datetime"}, inplace=True);
X_test.rename(columns={"reportts": "datetime"}, inplace=True);
y_train.rename(columns={"reportts": "datetime"}, inplace=True);

dataset = X_train.merge(y_train, on=['acnum', 'pos', 'datetime']).dropna(subset=['egtm'])
```

```python
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import mean_squared_error, mean_absolute_error

def train_model(X, y):
  assert len(X) == len(y)

  X = X.drop(columns=[
    'datetime', 'pos', 'fltdes'
  ])

  # обучаем на более ранних данных, тестируем на поздних
  split_index = int(len(X) * 0.75)

  X_train, y_train = X.iloc[:split_index], y.iloc[:split_index]
  X_val, y_val = X.iloc[split_index:], y.iloc[split_index:]


  model = CatBoostRegressor(
        depth=4,
        iterations=3500,
        learning_rate=0.06,
        loss_function="RMSE",
        
        cat_features=cat_features,
        # ignored_features = ignored_features,
        
        # Регуляризация и ускорение
#         colsample_bylevel=0.098,
#         subsample=0.95,
#         l2_leaf_reg=9,
#         min_data_in_leaf=243,
#         max_bin=187,
#         random_strength=1,
        
        # Параметры скорения
        task_type="CPU",    
        thread_count=-1,
        bootstrap_type="Bernoulli", 
        
        # Важное!
        random_seed=42,
        early_stopping_rounds=50)
    
#   model.fit(
#         train_dataset,
#         eval_set=eval_dataset,
#         verbose=500,
#         use_best_model=True,
#         plot=False)
    
  model.fit(X_train, y_train, 
            eval_set=(X_val, y_val), 
            early_stopping_rounds=50, 
            verbose=False,
            plot=False)

  predicted = model.predict(X_val)
  rmse = mean_squared_error(y_val, predicted, squared=False)
  mae = mean_absolute_error(y_val, predicted)

  return rmse, mae, model
```

```python
fleet = ['VQ-BGU', 'VQ-BDU']
models = []

for acnum in fleet:
  X = dataset_time_sorted[dataset_time_sorted['acnum'] == acnum].drop(columns=['egtm'])
  y = dataset_time_sorted[dataset_time_sorted['acnum'] == acnum]['egtm']

  rmse, mae, model = train_model(X, y)
  models.append(model)

  print(f'acnum={acnum} RMSE={rmse:.3f} MAE={mae:.3f}')
```

```python
tscv = TimeSeriesSplit(n_splits=3)
X = dataset[dataset['acnum'] == 'VQ-BGU'].drop(columns=['egtm'])
y = dataset[dataset['acnum'] == 'VQ-BGU']['egtm']
Xt1 = pd.DataFrame()
for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    print('-----------------')
    print(train_index, test_index)
```

```python
cat_features
```

```python
n_splits = 3  # Число фолдов. Оптимально 3, 5 или 10
models = [] # Тут будем сохранять модели
scores = [] # Тут будем хранить скоры валидационных фолдов

X = dataset_time_sorted[dataset_time_sorted['acnum'] == 'VQ-BGU'].drop(columns=['egtm','datetime', 'pos'])
y = dataset_time_sorted[dataset_time_sorted['acnum'] == 'VQ-BGU']['egtm']


tscv = TimeSeriesSplit(n_splits=n_splits)
for train_index, test_index in tscv.split(X):

    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = CatBoostRegressor(
        depth=4,
        iterations=3500,
        learning_rate=0.06,
        loss_function="RMSE",
        
        cat_features=cat_features,
        # ignored_features = ignored_features,
        
        # Регуляризация и ускорение
#         colsample_bylevel=0.098,
#         subsample=0.95,
#         l2_leaf_reg=9,
#         min_data_in_leaf=243,
#         max_bin=187,
#         random_strength=1,
        
        # Параметры скорения
        task_type="CPU",    
        thread_count=-1,
        bootstrap_type="Bernoulli", 
        
        # Важное!
        random_seed=42,
        early_stopping_rounds=50)

    models.append(model)

    model.fit(
        X_train, y_train, 
            eval_set=(X_test, y_test), 
            early_stopping_rounds=50, 
            verbose=False, plot=False)

    scores.append(np.mean([v for k, v in model.best_score_["validation"].items() if "RMSE" in k], dtype="float16"))
    # scores.append(clf.best_score_['validation']['MultiClass'])
    # clf.save_model("../tmp_data/catboost_model_{}".format(n))

assert len(models) == n_splits
print("mean/std of RMSE scores --------->", np.mean(scores, dtype="float16"), np.std(scores, dtype="float16"))
```
