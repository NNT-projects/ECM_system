```python
import pandas as pd
```

```python
X_train = pd.read_csv('X_train.csv', parse_dates=['reportts'])
y_train = pd.read_csv('y_train.csv', parse_dates=['reportts'])
```

```python
X_train.rename(columns={"reportts": "datetime"}, inplace=True)
y_train.rename(columns={"reportts": "datetime"}, inplace=True)
```

```python
print(X_train.shape,y_train.shape)
```

```python
df = pd.concat([X_train, y_train["egtm"]], axis=1)
df
```

```python
pd.unique(df['pos'])
```

```python
df = df.sort_values(by='datetime').reset_index(drop=True)
```

```python
df = df.dropna(axis=1, how="all")
df
```

```python
print(df['datetime'][(df['datetime'] > df['datetime'][5]) * (df['datetime'] < df['datetime'][100])])
print(df['datetime'][5])
print(df['datetime'][100])
# df['datetime'][0]
df["datetime"].min().date()
df.columns
```
