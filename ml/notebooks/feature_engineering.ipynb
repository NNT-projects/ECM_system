```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (12, 12)
```

```python
X_train = pd.read_csv('../data/X_train.csv', parse_dates=['reportts'])
y_train = pd.read_csv('../data/y_train.csv', parse_dates=['reportts'])

X_test = pd.read_csv('../data/X_test.csv', parse_dates=['reportts'])


dataset = X_train.merge(y_train, on=['acnum', 'pos', 'reportts']).dropna(subset=['egtm'])
```

```python
# Delete cols where all values are Nan
dataset = dataset.dropna(axis=1, how='all')

# Fill NaNs
dataset = dataset.fillna(method='ffill')
```

```python
dataset_time_sorted = dataset.sort_values(by='reportts').reset_index(drop=True)
```

```python
dataset_time_sorted.info(verbose=True)
```

## Генерируем "скользящие" статистики

```python
data = dataset_time_sorted.copy()
```

```python
data['acnum'] = data['acnum'].apply(lambda x: 3 if 'VQ-BGU' else 4)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel



# Define a list of features to generate rolling statistics for
feature_list = data.columns.difference(['reportts', 'acnum', 'pos', 'fltdes', 'dep', 'arr', 'egtm'])

# Generate rolling statistics
window_size = 5

for feature in feature_list:
    data[f'{feature}_rolling_mean'] = data.groupby(['acnum', 'pos'])[feature].rolling(window=window_size, min_periods=1).mean().reset_index(level=[0,1], drop=True)
    data[f'{feature}_rolling_std'] = data.groupby(['acnum', 'pos'])[feature].rolling(window=window_size, min_periods=1).std().reset_index(level=[0,1], drop=True)
    data[f'{feature}_rolling_min'] = data.groupby(['acnum', 'pos'])[feature].rolling(window=window_size, min_periods=1).min().reset_index(level=[0,1], drop=True)
    data[f'{feature}_rolling_max'] = data.groupby(['acnum', 'pos'])[feature].rolling(window=window_size, min_periods=1).max().reset_index(level=[0,1], drop=True)

# Separate features and target
X = data.drop(columns=['egtm', 'reportts', 'acnum', 'pos', 'fltdes', 'dep', 'arr'])
y = data['egtm']

# Use Random Forest for feature selection
model = RandomForestRegressor(n_estimators=100)
model.fit(X, y)

# Select features based on importance
selector = SelectFromModel(model, prefit=True)
X_selected = selector.transform(X)

# Display selected features
selected_features = X.columns[selector.get_support()]
print("Selected features:", selected_features)
```

```python
# Filter the selected features to include only the rolling features
rolling_features = [feature for feature in selected_features if 'rolling' in feature]

# Create a DataFrame with the selected rolling features
selected_rolling_features_df = data[rolling_features]

# Concatenate the selected rolling features with the original dataset
dataset_with_selected_rolling_features = pd.concat([dataset_time_sorted, selected_rolling_features_df], axis=1)

# Display the first few rows to verify
dataset_with_selected_rolling_features.head()

dataset = dataset_with_selected_rolling_features.copy()
```

## Генерируем параметры из datetime

```python
# Convert the 'reportts' column to datetime format
dataset['reportts'] = pd.to_datetime(dataset['reportts'])

# Extracting various time-based features
dataset['year'] = dataset['reportts'].dt.year
dataset['month'] = dataset['reportts'].dt.month
dataset['day'] = dataset['reportts'].dt.day
dataset['day_of_week'] = dataset['reportts'].dt.dayofweek
dataset['hour'] = dataset['reportts'].dt.hour
dataset['minute'] = dataset['reportts'].dt.minute
dataset['second'] = dataset['reportts'].dt.second
```

```python
# Determine the part of the day
def get_part_of_day(hour):
    if 5 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 17:
        return 'afternoon'
    elif 17 <= hour < 21:
        return 'evening'
    else:
        return 'night'


# Determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'winter'
    elif month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    else:
        return 'fall'

```

```python
dataset['part_of_day'] = dataset['hour'].apply(get_part_of_day)
dataset['season'] = dataset['month'].apply(get_season)
```

```python
dataset.to_csv('../tmp_data/dataset_with_new_features.csv', index=False)
```
